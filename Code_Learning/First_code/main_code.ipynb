```python
import pandas as pd

# Base API URL
API_URL = "https://data.ct.gov/resource/rybz-nyjw.csv"

# Parameters
LIMIT = 1000  # Max records per request
OFFSET = 0    # Start offset
DATAFRAMES = []  # Store each chunk

while True:
    # Construct paginated URL
    paginated_url = f"{API_URL}?$limit={LIMIT}&$offset={OFFSET}"
    
    # Read CSV chunk
    df_chunk = pd.read_csv(paginated_url)
    
    # Break if no more data
    if df_chunk.empty:
        break
    
    # Append to list
    DATAFRAMES.append(df_chunk)
    
    # Update offset
    OFFSET += LIMIT
    print(f"Fetched {OFFSET} records...")

# Combine all chunks
df_full = pd.concat(DATAFRAMES, ignore_index=True)

# Save to CSV
df_full.to_csv("Accidental_Drug_Deaths_2012_2023.csv", index=False)

```
